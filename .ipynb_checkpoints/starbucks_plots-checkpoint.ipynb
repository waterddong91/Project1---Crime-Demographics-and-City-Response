```python
# Dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as st
```

```python
edited_final_df = pd.read_csv('Edited_Final_Data.csv')
edited_final_df.drop('Unnamed: 0', axis='columns', inplace=True)
edited_final_df
```

## Linear Regression Plots

```python
# Create a function to calculate regression values and creat linear regression plots
def linear_regression_plot(x_values, y_values, x_label, y_label, annotate_point):
    
    mask = ~np.isnan(x_values) & ~np.isnan(y_values)
    slope, intercept, r, p, std_err = st.linregress(x_values[mask], y_values[mask])
    
    # Regression values
    regression_values = slope * x_values + intercept
    
    # Create a equation of line    
    eq_line = f"{round(slope, 2)}x + {round(intercept, 2)}"

    # Create a plot
    plt.scatter(x_values, y_values, color='green', alpha=0.5)
    
    plt.plot(x_values, regression_values, color='green')
    plt.annotate(eq_line, annotate_point, color='green', fontsize=16)
    
    plt.title(f'{y_label} vs. {x_label}')
    plt.xlabel(f'{x_label}')
    plt.ylabel(f'{y_label}')
               
    plt.grid()

    # Save the figures
    # plt.savefig(f"output_data/{y_label}_vs_{x_label}-linear_regression.png")
    plt.show()
    
    print(f'The correlation coefficient is {r}.')
    print(f'p value is {p}.')
    print(f'r square value is {r**2}.')
```

## With the whole dataset (not excluding outliers)

```python
# final_df variables (did not exclude outliers)

final_avg_income = edited_final_df['Average_Income']
final_sb = edited_final_df['Number of Starbucks Stores']
final_crime = edited_final_df['Total Average Rate']
final_home_prices = edited_final_df['Home Prices']
final_population = edited_final_df['Population, 2016']

starbucks_label = 'Total Number of Starbucks Stores'
avg_income_label = 'Average Income'
med_income_label = 'Median Income'
home_prices_label = 'House Prices'
population_label = 'Population'
crime_label = 'Total Average Crime Rate'
```

### Average Income vs. Starbucks Stores

```python
# final_df - average income vs. starbucks stores

linear_regression_plot(final_avg_income, final_sb, avg_income_label, starbucks_label, (200000, 15))
```

### Crime Rate vs. Starbucks Stores

```python
# final_df - crime rate vs. starbucks stores

linear_regression_plot(final_crime, final_sb, crime_label, starbucks_label, (500, 15))
```

### Home prices vs. Starbucks Stores

```python
linear_regression_plot(final_home_prices, final_sb, home_prices_label, starbucks_label, (400000, 10))
```

### Population vs. Starbucks Stores

```python
linear_regression_plot(final_population, final_sb, population_label, starbucks_label, (10000, 10))
```

## Without Average Income Outliers

```python
# Calculate the IQR and quantitatively determine if there are any potential outliers in Average Income.

# Calculate qualtiles and IQR
qualtiles = edited_final_df['Average_Income'].quantile([0.25, 0.5, 0.75])

first_qualtile = qualtiles[0.25]
third_qualtile = qualtiles[0.75]

iqr = third_qualtile - first_qualtile

# Calculate lower bound and upper bound
lower_bound = first_qualtile - (1.5 * iqr)
upper_bound = third_qualtile + (1.5 * iqr)

# Determine outliers using upper and lower bounds
outlier_df = edited_final_df.loc[(edited_final_df['Average_Income'] > upper_bound) | (edited_final_df['Average_Income'] < lower_bound), :]

outlier_df.head()
```

```python
# The final dataframe without average income outliers
outliers_removed_df = edited_final_df.loc[(edited_final_df['Average_Income'] <= upper_bound) & (edited_final_df['Average_Income'] >= lower_bound), :]

outliers_removed_df
```

```python
# outlier_removed_df variables
cleaned_avg_income = outliers_removed_df['Average_Income']
cleaned_sb = outliers_removed_df['Number of Starbucks Stores']
cleaned_med_income = outliers_removed_df['Median_Income']
```

### Average Income vs. Starbucks Stores (without average income outliers)

```python
linear_regression_plot(cleaned_avg_income, cleaned_sb, avg_income_label, starbucks_label, (200000, 15))
```
